```python
import pandas as pd

df = pd.read_csv("~/Datasets/hack1robo/data.csv", index_col=0)
X = df.drop("chapter",axis=1).select_dtypes(include=int)
```

```python
import numpy as np
from sklearn.decomposition import PCA, KernelPCA
from sklearn.metrics.pairwise import laplacian_kernel
import seaborn as sns

pca = PCA()
print(X.shape)
S = pca.fit_transform(X.values)
pca_df = pd.DataFrame(S, columns=[f"pca_{i}" for i in range(S.shape[1])])
sns.scatterplot(pca_df, x="pca_1", y="pca_2", alpha=0.4)
```
